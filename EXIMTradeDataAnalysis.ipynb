Now we find annual growth for each 1 year period between 2015-2019. This will provide us an essence of how did the product fare in the given duration with its peers. 
Growth formula is generalised {(Final Value/Initial Value)-1}* 100 and is in percentages. 

```python
df1['Growth % 15-16'] = ((df1['Exported value in 2016']/df1['Exported value in 2015'])-1)*100
```

```python
df1['Growth % 16-17'] = ((df1['Exported value in 2017']/df1['Exported value in 2016'])-1)*100
```

```python
df1['Growth % 17-18'] = ((df1['Exported value in 2018']/df1['Exported value in 2017'])-1)*100
```

```python
df1['Growth % 18-19'] = ((df1['Exported value in 2019']/df1['Exported value in 2018'])-1)*100
```

CAGR, more frequently used in Finance for calculating investment returns, abbreviated for Compounded Annual Growth Rate is a measure which gives a more realistic figure for growth of investment across a period. It is given by the formula,

CAGR = [{(F/I)^(1/T)}-1]*100   where,

F --> Final Value
I --> Initial Value
T --> Time period of observation

```python
df1['CAGR 2015-19']= ((df1['Exported value in 2019']/df1['Exported value in 2015'])**(1/4.0)-1)*100
```

```python
df1.drop(['Code'],axis=1)
```

```python
df1=df1.drop(['Code'],axis=1)
```

```python
df1=df1.sort_index()
```

```python
df1['Average growth 2015-19']=((df1['Exported value in 2019']/df1['Exported value in 2015'])-1)*100
```

```python
df1=df1[df1['CAGR 2015-19']>0]
```
